```python
# Identify numerical columns (float or int types)
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
print("Numerical Columns:", numerical_cols)

# Fill missing values in numerical columns with the mean
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())

# Identify categorical columns (object type)
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_cols)

# Fill missing values in categorical columns with the mode (most frequent value)
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print(df.describe())
```

```python
# Select only numerical columns (exclude categorical ones)
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate correlation only for numerical columns
correlation = df[numerical_cols].corr()
print(correlation)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

# Now calculate correlation
correlation = df.corr()
print(correlation)
```

```python
time_columns = ['Study_Hours_Per_Day', 'Extracurricular_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day']
avg_time = df[time_columns].mean()

plt.figure(figsize=(8, 6))
plt.bar(time_columns, avg_time, color=['blue', 'orange', 'green', 'red', 'purple'])
plt.title('Average Time Spent Per Activity')
plt.xlabel('Activity')
plt.ylabel('Average Hours Per Day')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group and calculate average social hours for sleep hours
social_vs_sleep = df.groupby('Sleep_Hours_Per_Day')['Social_Hours_Per_Day'].mean()

# Plot a line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=social_vs_sleep.index, y=social_vs_sleep.values, marker='o', color='orange')
plt.title('Social Hours vs Sleep Hours', fontsize=14)
plt.xlabel('Sleep Hours Per Day', fontsize=12)
plt.ylabel('Average Social Hours Per Day', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create bins for Study Hours and Sleep Hours
df['Study_Hours_Binned'] = pd.cut(df['Study_Hours_Per_Day'], bins=5)
df['Sleep_Hours_Binned'] = pd.cut(df['Sleep_Hours_Per_Day'], bins=5)

# Group by bins and calculate mean Stress Level
stacked_data = df.groupby(['Study_Hours_Binned', 'Sleep_Hours_Binned'])['Stress_Level'].mean().unstack()

# Plot stacked bar chart
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6), cmap='coolwarm', edgecolor='black')
plt.title('Stress Levels Across Study and Sleep Hours', fontsize=14)
plt.xlabel('Study Hours Range', fontsize=12)
plt.ylabel('Stress Level', fontsize=12)
plt.legend(title='Sleep Hours Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix for stress-related factors
stress_corr = df[['Study_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day',
                  'Extracurricular_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day', 'Stress_Level']].corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(stress_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation of Factors with Stress Level', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
time_columns = ['Study_Hours_Per_Day', 'Extracurricular_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Social_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day']
avg_time = df[time_columns].mean()

plt.figure(figsize=(8, 8))
plt.pie(avg_time, labels=time_columns, autopct='%1.1f%%', startangle=140)
plt.title('Average Time Distribution Among Activities')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate GPA based on study hours
study_hours_avg_gpa = df.groupby('Study_Hours_Per_Day')['GPA'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=study_hours_avg_gpa.index, y=study_hours_avg_gpa.values, marker='o', color='blue')
plt.title('GPA Trends Across Study Hours', fontsize=14)
plt.xlabel('Study Hours Per Day', fontsize=12)
plt.ylabel('Average GPA', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

INSIGHTS

*  Both insufficient and excessive sleep negatively impact GPA. Students who sleep between 6-8 hours tend to perform better.
*  High stress levels are inversely correlated with GPA. Students with lower stress levels perform better academically.
*  A moderate amount of social interaction supports better performance, but excessive socializing lowers GPA.
*  Physical activity has minimal direct impact on GPA but helps reduce stress and improves overall mental health.

CONCLUSION

Promote a balanced sleep schedule to ensure students are well-rested and Provide mental health support, time management training, and relaxation techniques.
